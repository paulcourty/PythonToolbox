## Colors

```python
import matplotlib.pyplot as plt

colors = ['r', 'b', 'g', 'k', 'mediumaquamarine', 'c', 'm', 'y', 'midnightblue'] 

x_initial = [0]
y_initial = [0]
size = 1000


scatter, ax = plt.subplots(figsize = (24, 4))

for count, color in enumerate(colors):
    ax.scatter(x_initial[0] + count, y_initial, s = size, c = color)
    
ax.set_xlim([-0.5, len(colors) - 0.5])
ax.axis('off')
plt.show()
```

Note: Example set of colors, find all @ https://matplotlib.org/stable/gallery/color/named_colors.html

## Plots

```python
N = 10
x = np.sort(np.random.rand(N))
y = np.random.rand(N)


# PLOT
fig, ax = plt.subplots(figsize = (18, 6))

ax.plot(x, y, label = 'ex', color = 'k')
ax.legend()
ax.set_title('')
ax.set_xlabel('')
ax.set_ylabel('')

plt.show()
```

---

## Subplots

```python
N = 10
x1 = np.sort(np.random.rand(N))
y1 = np.random.rand(N)
x2 = np.sort(np.random.rand(N))
y2 = np.random.rand(N)


# PLOT
fig, axs = plt.subplots(2, 1, figsize = (18, 13))

axs[0].plot(x1, y1, label = 'ex', color = 'k')
axs[0].legend()
axs[0].set_title('')
axs[0].set_xlabel('')
axs[0].set_ylabel('')
axs[0].set_xlim()
axs[0].set_ylim()

axs[1].plot(x2, y2, label = 'ex', color = 'k')
axs[1].legend()
axs[1].set_title('')
axs[1].set_xlabel('')
axs[1].set_ylabel('')
axs[1].set_xlim()
axs[1].set_ylim()

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
plt.show()
```

## Bar Plots

```python
N = 10
x = np.random.randint(low = 0, high = 10, size = N)
height = np.random.rand(N)


# PLOT
bar, ax = plt.subplots(figsize = (18, 6))

ax.bar(x, height, color = 'mediumaquamarine')
ax.set_title('')
ax.set_xlabel('')
ax.set_ylabel('')

plt.show()
```

---

## Scatter Plots

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii


# PLOT
scatter, ax = plt.subplots(figsize = (18, 6))

ax.scatter(x, y, s=area, c=colors, alpha=0.5)
ax.set_title('')
ax.set_xlabel('')
ax.set_ylabel('')

plt.show()
```

---

## Pie Charts

```python
N = 3
x = np.random.rand(N) + 1
labels = ['label_1', 'label_2', 'label_3']


# PLOT
pie, ax = plt.subplots(figsize = (10, 8))

plt.pie(x, autopct = "%.1f%%", explode = len(x) * [0.01], labels = labels, pctdistance = 0.5, textprops = {'fontsize': 15})
plt.title('', fontsize = 14)
plt.show()
```
