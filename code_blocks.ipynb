# Machine Learning

**Test Dataset**: Iris Flowers

```python
from sklearn import datasets

iris = datasets.load_iris()

X = iris.data
y = iris.target

print('\nDataset: Num. of flowers {} and Num. of features {}'.format(X.shape[0], X.shape[1]))
```

## Split Train / Test

**Note**: Dataset format for Sklearm should be _(data points, features)_

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

## Classification | Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state = 0)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
acc = clf.score(X_test, y_test)
pred_prob = np.round(clf.predict_proba(X_test), 2)

print('\nAccuracy:', np.round(100 * acc, 1))

print('\nExample of Prediction Probabilities for each class for first 3 elements of X_test:\n')
for i in range(3):
    print('   Element {}:  C1   {}  ,  C2   {}  ,  C3   {}'.format(i, pred_prob[i][0], pred_prob[i][1], pred_prob[i][2]))
```

## Classification | Knn

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Knn example for arbitrary K and weights 'distance' (pretty interesting parameter)
k_neighb = 30
knn = KNeighborsClassifier(n_neighbors = k_neighb, weights = 'distance')
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
acc = knn.score(X_test, y_test)

print('\nAccuracy:', np.round(100 * acc, 1))
```

## Confusion Matrix

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
labs = ['Class 1', 'Class 2', 'Class 3']

fig = plt.figure(figsize = (14, 6))

plt.matshow(confusion_matrix(y_test, y_pred, normalize = 'true'), cmap = 'Blues')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix | Knn (k = ' + str(k_neighb) + ') | Acc ' + str(np.round(100 * acc, 1)) + '%\n')
for (i, j), z in np.ndenumerate(confusion_matrix(y_test, y_pred, normalize = 'true')):
    plt.text(j, i, '{:0.2f}'.format(z), ha='center', va='center', bbox=dict(boxstyle='round', facecolor='white', edgecolor='1'))

plt.show()
```

## Clustering | Kmeans

```python
from sklearn.cluster import KMeans
```

```python
# Kmeans example
kmeans = KMeans(n_clusters = 3, random_state = 42)
y_pred = kmeans.fit_predict(X)

clusters = kmeans.cluster_centers_
```

```python
# PLOT EXAMPLE FOR 3 FEATURES
# Actual
print('\nActual Clusters | Example for 3 / 4 Features\n')

fig, axs = plt.subplots(1, 3, figsize = (20, 4))

axs[0].scatter(X[:, 0], X[:, 1], c = y, s = 50, cmap = 'viridis')
axs[1].scatter(X[:, 0], X[:, 2], c = y, s = 50, cmap = 'viridis')
axs[2].scatter(X[:, 1], X[:, 2], c = y, s = 50, cmap = 'viridis')

plt.show()


# Created / Predicted Clusters
print('\n\n\nKmeans Clusters for 3 / 4 Features\n')

fig, axs = plt.subplots(1, 3, figsize = (20, 4))

axs[0].scatter(X[:, 0], X[:, 1], c = y_pred, s = 50, cmap = 'viridis', alpha = 0.5)
axs[0].scatter(clusters[:, 0], clusters[:, 1], c = 'r', s = 200, marker = 'x', linewidths = 3)
axs[1].scatter(X[:, 0], X[:, 2], c = y_pred, s = 50, cmap = 'viridis', alpha = 0.5)
axs[1].scatter(clusters[:, 0], clusters[:, 2], c = 'r', s = 200, marker = 'x', linewidths = 3)
axs[2].scatter(X[:, 1], X[:, 2], c = y_pred, s = 50, cmap = 'viridis', alpha = 0.5)
axs[2].scatter(clusters[:, 1], clusters[:, 2], c = 'r', s = 200, marker = 'x', linewidths = 3)


plt.show()
```

## PCA

```python
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# PCA example
# PCA 3 dimensions
fig = plt.figure(1, figsize = (8, 6))
ax = Axes3D(fig, elev = -155, azim = 115, auto_add_to_figure = False)

X_reduced = PCA(n_components = 3).fit_transform(iris.data)

ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c = y, cmap = plt.cm.Set1, s = 50)
ax.set_title("\nPCA 3 directions", fontsize = 16)
ax.set_xlabel("\n1st eigenvector")
ax.set_ylabel("\n2nd eigenvector")
ax.set_zlabel("\n3rd eigenvector")

fig.add_axes(ax)
plt.show()


# PCA 2 dimensions
fig = plt.figure(2, figsize = (8, 6))

X_reduced = PCA(n_components = 2).fit_transform(iris.data)

plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c = y, cmap = plt.cm.Set1, s = 50)
plt.title("\nPCA 2 directions\n", fontsize = 16)
plt.xlabel("\n1st eigenvector")
plt.ylabel("\n2nd eigenvector")

plt.show()
```

## PostgreSQL

```python
import psycopg2 as pg2
```

Create Database

```python
def create_database(database_name, user_name, user_password):
    try:
        conn = pg2.connect(database = 'postgres', user = user_name, password = user_password)
        conn.autocommit = True
        cur = conn.cursor()
        cur.execute('CREATE DATABASE ' + database_name)
        conn.close()
        print('\nDatabase created\n')
    except:
        print('\nDatabase already exists\n')
```

## PostgreSQL

```python
import sqlite3
```

Query to Dataframe

```python
def query_to_df(database, query):
    conn = sqlite3.connect(database)  
    c = conn.cursor()
    c.execute(query)
    
    fields = [description[0] for description in c.description]
    query_df = pd.DataFrame(c.fetchall(), columns = fields)
    
    conn.close()
    
    return query_df
```
